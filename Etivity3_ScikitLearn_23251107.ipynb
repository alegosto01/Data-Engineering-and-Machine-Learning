```python
# import dependencies
import numpy as np
```

1.	Read the iris dataset using Pandas (ref Option 2 above)
2.	Extract the Iris-setosa and Iris-verginica samples only
3.	Extract the petal length and petal width. These shall be our features.

```python
import pandas as pd
df = pd.read_csv('iris_data.csv', header=0)

#Extracting setosa and virginica, choosing everything besides versicolor
df = df.loc[df[' class']!='Iris-versicolor']

#Extracting petal length and petal width
X = df.iloc[:, [2, 3]].values

#Extracting the class column mapping the two classes into 1 or -1 value
y = df.iloc[:, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)
```

4.	Plot the dataset as a scatter plot with the two Iris types shown as different colours

```python
import matplotlib.pyplot as plt

#Plotting petal classification

plt.scatter(X[:,0], X[:,1],c=y)
plt.title("Scatter plot of petal sizes")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show() 
```

5.	Use the perceptron in sklearn.linear_model to train on the Iris data previously used. Set verbose = 1; have a look at the epochs, it converges very quickly?

```python
from sklearn.linear_model import Perceptron as Perceptron 

#Defining the classificator and training it on the data
clf = Perceptron(verbose=1)
clf.fit(X,y)
```

The perceptron stops after 7 epochs,but we can see that the average loss at the second epoch is already 0.0 . It converged very quickly due to the fact that the two classes have really different features' values that make them very easy to classificate

6.	Using many random inputs (e.g. 5000) plot to see if you can see the decision boundary for the feature scaled case

```python
import random

n_inputs = 5000
i = 0
petal_length = []
petal_width = []


#Creating 5000 random sizes 
petal_length = np.random.uniform(min(X[:,0]), max(X[:,0]),n_inputs)
petal_width = np.random.uniform(min(X[:,1]), max(X[:,1]), n_inputs)

#Stacking the arrays in to a matrix with the right sizes
random_data = np.column_stack((petal_length, petal_width))

#Predict the class of each random petal
random_y_pred = clf.predict(random_data)

#Plotting the random iris
plt.scatter(petal_length, petal_width, c=random_y_pred,s=10)
plt.title("Random iris sizes plot")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show() 
```

Similarities:

-Workflow: Both the neurons have the same workflow since the artificial ones are inspired by the biological ones. They both get information from one “side”, process the information and send the processed data to the next neuron. 

-Signals: Connections can become stronger or weaker in biological systems and this behavior can be simulated in artificial neurons increasing and decreasing the weights.

-Learning: Both types of neurons learn by their input. Humans improve their abilities learning by previous experience, the same as artificial neural networks do   through gradient descent and backpropagation

-Activation function: In both environments, we find a weighted sum of the inputs. In the biological function, the sum triggers or not the neurons depending on passing or not the threshold, while in artificial neurons we have mathematical functions that calculate a floating point output based on the weighted sum.	

Differences:

-Adaptability: One of the main differences between humans and models is the fact that the model needs to learn the "rules" of the problem to become very good at solving it. If we change a detail of the problem the model could really decrease drastically its performances. Biological systems instead, are very good at adapting previous experience to new slightly different problems and performing well at them

-Size: In our brain, we have almost 90 billion neurons and more than 100 trillion connections between them. The number of neurons in the artificial networks is much lower than this. In artificial neuron networks usually, we have a first layer of neurons, as many as the input features, and a final layer made by as many neurons as many classes we want to predict if they are built for a classification problem. In between these layers, there are several hidden layers fully connected: the first with the second, the second with the third, and so on. The limitation of the size of artificial networks is not only computational but mostly functional. Increasing the number of neurons does not always bring you better performance.

-Workflow: In artificial neuron networks all layers are computed one by one, synchronously in only one direction, besides the backpropagation algorithm that can change weights in the opposite way. Moreover, each layer is connected to the previous and the next but not with all the others. In the biological networks instead, neurons work asynchronously, in some small parts of highly connected neurons and many large parts with fewer connections. The only way to simulate this in artificial neurons is to set a weight to 0 to turn off the connection between neurons.

-Speed: in biological neurons the speed depends on the person’s characteristics and can be affected by fatigue while artificial ones have no refractory period as biological ones have and can be highly optimized with better hardware and better math calculations.

-Fault-tolerance: Biological networks are fault-tolerant, so the information stored is redundant, while artificial ones don’t suffer from this problem since training for a long time doesn’t affect efficiency.

-Signals: Unlike biological neurons, artificial neurons don’t use binary signals. Instead, continuous values are used as weight, allowing them to be used in many mathematical applications. 

-Learning: in biological systems, repetition is the key to learning anything. In contrast in an artificial environment once information is stored can’t be forgotten as humans do. This creates an error rate in biological networks. A human brain fed with the same input can give you different outputs; it can make mistakes. In contrast, artificial neuron networks will always give you the same output from the same input.
